```python
# problem 1
import numpy as np
# parameters of dynamics
mu = 0.5
sigma = 0.1
r = 1. #0
p = 0.005
T = 0.1
dt = 0.001

# theoretical v-function and control_a:
theo_v_list = []
theo_a_list = []
for i in [2,5,10,15,20]:
    a = (mu-r)/((sigma**2)*(1.-p))
    rho = p*((mu-r)**2)/(2.*(sigma**2)*(1.-p)) + r*p
    v = np.exp(rho*T)*(i**p/p)
    v_theo = np.round(v,2)
    a_theo = np.round(a,2)
    
    theo_v_list.append(v_theo)
    theo_a_list.append(a_theo)
print(theo_v_list)
print(theo_a_list)
    
```
