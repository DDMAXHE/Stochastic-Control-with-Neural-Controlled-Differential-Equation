```python
import numpy as np
# parameters of dynamics
sigma = 0.1
# those parameters can be vectors
A_t = 0
sigma_t = sigma
B_t = 1
Q_t = 0
Q_T = 1
R_t = 1
T = 0.1
dt = 0.001

# theoretical v-function and control_a:
theo_v_list = []
theo_a_list = []
for i in [2,5,10,15,20,50]:
    
#     a = (mu-r)/((sigma**2)*(1.-p))
#     rho = p*((mu-r)**2)/(2.*(sigma**2)*(1.-p)) + r*p
#     v = np.exp(rho*T)*(i**p/p)
#     v_theo = np.round(v,2)
#     a_theo = np.round(a,2)
    
    v = i**2*Q_T/(Q_T*T+1) + sigma**2*np.log(Q_T*T+1)
    P_0_star = Q_T/(Q_T*T+1)          
    K_t_star = R_t*B_t*P_0_star
    a_0_star = -K_t_star*i
    v_theo = np.round(v,2)
    a_theo = np.round(a_0_star,2)
    
    theo_v_list.append(v_theo)
    theo_a_list.append(a_theo)
print(theo_v_list)
print(theo_a_list)
    
```
